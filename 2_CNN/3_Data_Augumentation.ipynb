```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
test_img = cv2.imread('../../DSML_CodeSpyder/Deep_Learning/Images/cat.png')
```

```python
plt.imshow(test_img)
```

```python
test_img.shape
```

```python
test_img = cv2.resize(test_img,(224,224))
```

```python
test_input = test_img.reshape((1,224,224,3))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Augumentation
data_gen = ImageDataGenerator(rotation_range=.2,
                              shear_range=0.2,
                              horizontal_flip=True,
                              vertical_flip=True,
                              rescale=1/255,
                              width_shift_range=0.2,
                              height_shift_range=0.2
)
```

```python
# all created images are stored in aug folder.

i = 0
for output in data_gen.flow(test_input,batch_size=1,save_to_dir='aug'):
    i = i + 1
    if i == 10:
        break
```
